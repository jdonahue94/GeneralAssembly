```python
import pandas as pd
```

```python
# creating a simple dataframe from a dictionary
simple_dict = {letter: [i for i in range(3)] for letter in 'abc'}
print(simple_dict)
```

```python
pd.DataFrame(simple_dict)
```

```python
# a little something fancier
list_of_dicts = [{'a': 1, 'b': 2, 'c': 3} for i in range(5)]
list_of_dicts
```

```python
pd.DataFrame(list_of_dicts)
```
